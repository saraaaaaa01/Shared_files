```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

# Assuming df is your DataFrame with columns 'Sector' and 'Words'
# For example, you can load your data from a CSV file like this:
# df = pd.read_csv('your_data.csv')

# Define sectors and the number of elements per sector
sector_data = {
    'Agriculture, Sylviculture et Pêche': 131,
    'Barrages et Energie hydroélectrique': 5,
    'Centrales thermiques': 9,
    'Charbon thermique': 18,
    'Mines': 32,
    'NUCLÉAIRE CIVIL': 9,
    'Pétrole et Gaz': 13,
    'Tabac': 6,
}

# Assign a unique color to each sector
sector_colors = {sector: np.random.rand(3,) for sector in sector_data.keys()}

# Function to check if two circles overlap
def overlap(c1, c2):
    distance = np.sqrt((c2[0] - c1[0])**2 + (c2[1] - c1[1])**2)
    return distance < c1[2] + c2[2]

# Function to draw circles for each sector
def draw_circles(ax, sector_data):
    circles = []
    for i, (sector, count) in enumerate(sector_data.items()):
        # Calculate the diameter based on the number of words
        diameter = min(count * 0.3, 20)  # Adjusted scaling factor

        # Use polar coordinates to distribute clusters evenly
        angle = i * (2 * np.pi / len(sector_data))
        x_center = 1.5 * np.cos(angle)
        y_center = 1.5 * np.sin(angle)

        # Check for overlap and adjust positions if needed
        for c in circles:
            if overlap((x_center, y_center, diameter / 8), c):
                offset = np.array([2, 2])  # Adjust the relocation offset
                x_center += offset[0]
                y_center += offset[1]

        # Generate random points within the circle
        points = np.array([[np.random.uniform(-diameter / 8, diameter / 8) + x_center,
                             np.random.uniform(-diameter / 8, diameter / 8) + y_center] for _ in range(count)])

        # Apply KMeans clustering to the points
        if count > 1:
            kmeans = KMeans(n_clusters=min(count, 5), random_state=42)
            labels = kmeans.fit_predict(points)

            # Plot the clustered points
            for j in range(min(count, 5)):
                cluster_points = points[labels == j]
                plt.scatter(cluster_points[:, 0], cluster_points[:, 1], color=sector_colors[sector], alpha=0.7)

        # Draw the final circle
        final_circle = plt.Circle((x_center, y_center), diameter / 8, edgecolor='black', facecolor='none')
        ax.add_patch(final_circle)
        circles.append((x_center, y_center, diameter / 8))

# Create a scatter plot
plt.figure(figsize=(10, 8))
ax = plt.gca()

# Draw circles for each sector and scatter clustered points within initial circles
draw_circles(ax, sector_data)

# Add legend
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=sector) for sector, color in sector_colors.items()]
ax.legend(handles=handles, loc='upper right')

# Show the plot
plt.title('Word Clusters by Sector (Simulated Clustering with Closer Centers)')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()
```

```python
# Construct dictionary
sector_word_dict = {}

for _, row in df.iterrows():
    sector = row['Sector']
    word = row['Words']

    # Check if the sector is already a key in the dictionary
    if sector in sector_word_dict:
        # If yes, append the word to the existing list of words for that sector
        sector_word_dict[sector].append(word)
    else:
        # If no, create a new entry with the sector as the key and a list containing the word as the value
        sector_word_dict[sector] = [word]

# Print the constructed dictionary
sector_word_dict
```

```python
# Load pre-trained BERT model and tokenizer for French
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
model = BertModel.from_pretrained('bert-base-multilingual-cased')
```

```python
# Tokenize words using BERT tokenizer
tokenized_data = df["Words"].apply(lambda x: tokenizer(x, return_tensors="pt"))
```

```python
# Function to get BERT embeddings for a given phrase
def get_bert_embedding(phrase):
    with torch.no_grad():
        outputs = model(**phrase)
    return outputs.last_hidden_state.mean(dim=1).numpy()

# Get BERT embeddings for each tokenized phrase
embeddings = tokenized_data.apply(lambda x: get_bert_embedding(x))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Function to calculate cosine similarity
def calculate_cosine_similarity(embedding1, embedding2):
    return cosine_similarity(embedding1.reshape(1, -1), embedding2.reshape(1, -1))[0][0]
```

```python
threshold = 0.6  # Adjust the threshold based on your requirements

# Function to check if a word is similar to its sector
def is_word_related(word_embedding, sector_embedding, threshold):
    similarity_score = calculate_cosine_similarity(word_embedding, sector_embedding)
    return similarity_score >= threshold
```

```python
results_df = pd.DataFrame(columns=['Word', 'Sector', 'Related'])

# Iterate over sectors and associated words in the constructed dictionary
for sector, words in sector_word_dict.items():
    sector_embedding = get_bert_embedding(tokenizer(sector, return_tensors="pt"))

    for word in words:
        word_embedding = get_bert_embedding(tokenizer(word, return_tensors="pt"))
        related = is_word_related(word_embedding, sector_embedding, threshold)

        # Append result to results_df
        results_df = results_df.append({'Word': word, 'Sector': sector, 'Related': related}, ignore_index=True)
```
